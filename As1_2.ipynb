```python
import json
import re

# 读取 JSON 文件
file_path = '/content/drive/My Drive/CHC5904/test1/renamed_traditional_data.json'

with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# 定义一个函数，用于清洗文本
def clean_text(text):
    # 使用正则表达式移除非中文字符和标点符号
    cleaned_text = re.sub(r'[^\u4e00-\u9fa5]', '', text)  # 只保留中文字符
    return cleaned_text

# 对每个章节的正文进行清洗
for chapter in data:
    if 'Traditional' in chapter:  # 确保该字段存在
        chapter['Traditional'] = [clean_text(word) for word in chapter['Traditional']]

# 将清洗后的数据写入新文件
new_file_path = '/content/drive/My Drive/CHC5904/test1/cleaned_traditional_data.json'
with open(new_file_path, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)

print("正文已成功清洗，移除所有非中文字符！")
```

```python
import json
import re
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from wordcloud import WordCloud
from nltk import ngrams
from sklearn.decomposition import PCA
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
import nltk
import jieba

# 确保安装所需的库
!pip install nltk matplotlib wordcloud scikit-learn jieba

nltk.download('punkt')

# 下载繁体中文字型
!wget -O SourceHanSerifTW-VF.ttf https://github.com/adobe-fonts/source-han-serif/raw/release/Variable/TTF/Subset/SourceHanSerifTW-VF.ttf

# 加入字型檔
fm.fontManager.addfont('SourceHanSerifTW-VF.ttf')

# 设置字体
plt.rcParams['font.family'] = 'Source Han Serif TW VF'

# 读取 JSON 文件
file_path = '/content/drive/My Drive/CHC5904/test1/cleaned_traditional_data.json'
with open(file_path, encoding='utf-8') as f:
    data = json.load(f)

# 定义孙悟空及其别名
character = "孫悟空"
aliases = ["美猴王", "大圣", "行者"]

# 定义停用词表
stopwords = set([
    "的", "了", "是", "在", "和", "有", "就", "都", "也", "与", "来",
    "这", "那", "为", "对", "能", "要", "而", "是", "个", "一个",
    "我们", "你", "他", "她", "它", "所", "及", "以", "但", "或", "如果","道","我","也","又",
    "把","到","此","上","个","一","一个","去","却","等","著","不","怎么","叫","啊","得","不是","好","走","即","见","忽"
])

# 存储别名及其上下文
context_data = {alias: [] for alias in aliases}

# 遍历章节
for chapter in data:
    chapter_text = ''.join(chapter['Traditional'])  # 假设文本在'Traditional'中
    for alias in aliases:
        # 正则表达式查找别名
        pattern = re.compile(r'(.{0,20})(' + re.escape(alias) + r')(.{0,20})')
        matches = pattern.findall(chapter_text)
        for match in matches:
            # 记录别名及其上下文
            context_data[alias].append((match[0] + match[1] + match[2]).strip())

# 生成词云并可视化
for alias, contexts in context_data.items():
    text = ' '.join(contexts)

    # 使用jieba进行分词
    seg_list = jieba.cut(text)
    # 过滤停用词
    filtered_words = [word for word in seg_list if word not in stopwords]
    segmented_text = ' '.join(filtered_words)

    wordcloud = WordCloud(font_path='SourceHanSerifTW-VF.ttf', width=800, height=400).generate(segmented_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"{character} - {alias} Context Wordcloud")
    plt.show()

# n-gram
all_contexts = {}
for alias, contexts in context_data.items():
    all_text = ' '.join(contexts)

    # 使用jieba进行分词
    seg_list = jieba.cut(all_text)
    # 过滤停用词
    filtered_words = [word for word in seg_list if word not in stopwords]
    segmented_text = ' '.join(filtered_words)

    tokens = nltk.word_tokenize(segmented_text)

    # 生成 2-grams 和 3-grams
    bi_grams = list(ngrams(tokens, 2))
    tri_grams = list(ngrams(tokens, 3))
    all_contexts[alias] = (bi_grams, tri_grams)

    print(f"{character} 的别名 '{alias}' 的 2-grams:")
    print(bi_grams[:5])  # 输出前5个示例
    print(f"{character} 的别名 '{alias}' 的 3-grams:")
    print(tri_grams[:5])  # 输出前5个示例

```

```python
# 定义唐三藏及其别名
tang_sanzang_aliases = ["玄奘", "御弟", "三藏"]

# 定义猪八戒及其别名
zhu_bajie_aliases = ["呆子", "八戒", "老猪"]

# 存储唐三藏和猪八戒的上下文
tang_sanzang_context_data = {alias: [] for alias in tang_sanzang_aliases}
zhu_bajie_context_data = {alias: [] for alias in zhu_bajie_aliases}

# 遍历章节，获取唐三藏的上下文
for chapter in data:
    chapter_text = ''.join(chapter['Traditional'])  # 假设文本在'Traditional'中
    # 获取唐三藏的别名上下文
    for alias in tang_sanzang_aliases:
        pattern = re.compile(r'(.{0,20})(' + re.escape(alias) + r')(.{0,20})')
        matches = pattern.findall(chapter_text)
        for match in matches:
            tang_sanzang_context_data[alias].append((match[0] + match[1] + match[2]).strip())

    # 获取猪八戒的别名上下文
    for alias in zhu_bajie_aliases:
        pattern = re.compile(r'(.{0,20})(' + re.escape(alias) + r')(.{0,20})')
        matches = pattern.findall(chapter_text)
        for match in matches:
            zhu_bajie_context_data[alias].append((match[0] + match[1] + match[2]).strip())

# 生成唐三藏的词云并可视化
for alias, contexts in tang_sanzang_context_data.items():
    text = ' '.join(contexts)

    # 使用jieba进行分词
    seg_list = jieba.cut(text)
    # 过滤停用词
    filtered_words = [word for word in seg_list if word not in stopwords]
    segmented_text = ' '.join(filtered_words)

    wordcloud = WordCloud(font_path='SourceHanSerifTW-VF.ttf', width=800, height=400).generate(segmented_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"唐三藏 - {alias} Context Wordcloud")
    plt.show()

# 生成猪八戒的词云并可视化
for alias, contexts in zhu_bajie_context_data.items():
    text = ' '.join(contexts)

    # 使用jieba进行分词
    seg_list = jieba.cut(text)
    # 过滤停用词
    filtered_words = [word for word in seg_list if word not in stopwords]
    segmented_text = ' '.join(filtered_words)

    wordcloud = WordCloud(font_path='SourceHanSerifTW-VF.ttf', width=800, height=400).generate(segmented_text)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f"猪八戒 - {alias} Context Wordcloud")
    plt.show()
```
